FACE AND EYE DETECT

    STEPS:
        1) read and show image
        2) create classifier
        3) find 4 main points
        4) draw rectangle
        5) show result

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('me.jpg')
plt.imshow(img)
plt.axis('off')
plt.show()

# way 2 to show image
#cv2.imshow('Image', img)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

```python
# create classifier
clf_face = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
clf_eye = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
```

```python
# find points
def detect(img=img):
    face_points = clf_face.detectMultiScale(img)
    for (x,y,w,h) in face_points:
        cv2.rectangle(img, (x,y),(x+w,y+h),(0,255,255),5)
        
    eye_points = clf_eye.detectMultiScale(img)
    for (x,y,w,h) in eye_points:
        cv2.rectangle(img, (x,y), (x+w, y+h), (255,255,0),5)
    
    return img

result = detect(img)
plt.imshow(result)
plt.axis('off')
plt.show()
```

```python
# try on 2nd image
img2 = cv2.imread('1.jpg')
result = detect(img2)
plt.imshow(result)
plt.axis('off')
plt.show()
```

```python
# solve the problem
img2 = cv2.imread('1.jpg')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
result = detect(img2)
plt.imshow(result)
plt.axis('off')
plt.show()
```
